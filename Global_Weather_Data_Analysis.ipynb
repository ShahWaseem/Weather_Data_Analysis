# Plugin Zip Code and Country Name to collect forecast at present!

```python
from noaa_sdk import noaa

n = noaa.NOAA()
res = n.get_forecasts('10038', 'US', True)
for i in res:
    print(i)
```

```python
res = pd.DataFrame(res,columns=['number','startTime','isDaytime','temperature','temperatureUnit'], res['startTime'].astype('datetime64'))

res
```

```python
plot.title('')
plot.xlabel('day')
plot.ylabel('degrees F +/- from average')
plot.bar(res['startTime'], res['temperature'], color="blue")
plot.show()
```
